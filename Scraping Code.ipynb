```python
data = pd.read_csv("MLBTheShowData.csv")
player_name = data['players']
player_id = data['id']
```

```python
def scrape_players(player_id, player_name):
    url = "https://theshownation.com/mlb20/items/" + str(player_id)
    print(url)
    html = urlopen(url)
    bs = BeautifulSoup(html.read(), 'lxml')
    tables = bs.find_all("table",{"id":"table-trends"})
    stats_tables = pd.read_html(str(tables))
    
    price_stats = pd.DataFrame(stats_tables[0][:-6])
    price_stats['Player'] = player_name
    return(price_stats)

scrape_players("60aaa330b5aff236a9de5acf769faf38", "Mike Trout")  
```
